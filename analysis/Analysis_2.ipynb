```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import f
plt.figure()
X = np.array([[[i, 0, 0] for i in range(2007, 2011)] + [[i, 1, i] for i in range(2011, 2017)]]).reshape(10, 3) #dummy variable
#X
y = [-244000,-262000,-288000,-301000,-347000,-356000,-396000,-446000,-461000,-481000]
reg = LinearRegression().fit(X, y)
reg.score(X, y)
X_n = np.array(list(range(2007, 2017))).reshape(-1, 1)
X_1 = np.array([[i, 0, 0] for i in range(2007, 2011)])
X_2 = np.array([[i, 1, i] for i in range(2011, 2017)])
reg.predict(X_1)
X_n
plt.plot(X_n[:4,0], reg.predict(X_1), "--")
plt.plot(X_n[4:,0], reg.predict(X_2), "--")
plt.plot(X_n, y)

def rss(y, y_pred):
    return mean_squared_error(y, y_pred)
def chow(y, y_pred_all, y_pred_1, y_pred_2, N, k, breakpoint = 4):
    rss_p = rss(y, y_pred_all)
    rss_1 = rss(y[:breakpoint], reg.predict(X_1))
    rss_2 = rss(y[breakpoint:], reg.predict(X_2))
    num = (rss_p - rss_1 - rss_2)/(k)
    den = (rss_1 + rss_2)/(N-2*k)
    return num/den
reg_all = LinearRegression().fit(X_n, y)
y_pred_all = reg_all.predict(X_n)
y_pred_1 = reg.predict(X_1)
y_pred_2 =  reg.predict(X_2)
print(chow(y, y_pred_all, y_pred_1, y_pred_2, 10, 2))
f.ppf(0.95,2,6)
```
