```python
import pandas as pd
import os
```

```python
df = pd.read_csv("london_olympics_trends.csv",skiprows=2)
```

```python
df.columns = ["Date", "GoogleTrendsInterest"]
```

```python
df["GoogleTrendsInterest"].unique()
```

```python
def impute_zero(val):
    if val == "<1":
        return 0
    else:
        return float(val)/100 
```

```python
df["GoogleTrendsInterest"] = df["GoogleTrendsInterest"].apply(impute_zero)
```

```python
df.set_index("Date").plot(figsize=(12,5))
```

```python
df.to_csv("london_google_trends_clean.csv", index=False)
```
